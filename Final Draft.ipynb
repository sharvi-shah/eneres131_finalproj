## Making megaDF

After cleaning all the datasets individually and making sure country codes and names matched, we were ready to merge them into one big dataset. To start this effort, we dropped any data for any years that were not recorded in all four datasets. Next, we renamed the columns so they were consistent between all datasets. In addition to changing changing "Entity" to "Country Name" and "Code" to "Country Code", we also had to change the columns with years so all columns had years as integers, instead of strings.

```python
# comparing column names
print(GDP.columns)  
print(population.columns)
print(total_consumption.columns)
print(renewables_percentage.columns)

# comparing column data types
print(GDP.columns.map(type))
print(population.columns.map(type))
print(total_consumption.columns.map(type))
print(renewables_percentage.columns.map(type))

# dropping columns with historic data that isn't shared between all 4 datasetes
GDP.drop(GDP.columns[2:GDP.columns.get_loc('2000')], axis=1, inplace=True)
population.drop(population.columns[2:population.columns.get_loc('2000')], axis=1, inplace=True)

# renaming columns so they're consistent between all datasets
yrStrToInt = {str(yr):int(yr) for yr in range(2000, 2023)}
GDP.rename(columns=yrStrToInt, inplace=True)
population.rename(columns=yrStrToInt, inplace=True)

total_consumption.rename(columns={'Entity':'Country Name', 'Code':'Country Code'}, inplace=True)
renewables_percentage.rename(columns={'Entity':'Country Name', 'Code':'Country Code'}, inplace=True)
```

After all data matched up between the datasets, we melted all four so we each dataset would have a column for the "Country Name", "Country Code", "year", and the associated data. Finally, we could merge all the datasets. After merging, we also changed the units so all energy consumption data was in kWh, instead of TWh. This made understanding numbers for energy consumption per capita easier to understand.

```python
GDP = GDP.melt(id_vars=['Country Name', 'Country Code'], value_vars=range(2000, 2023), var_name='year', value_name='GDP')
population = population.melt(id_vars=['Country Name', 'Country Code'], value_vars=range(2000, 2023), var_name='year', value_name='population')
total_consumption = total_consumption.melt(id_vars=['Country Name', 'Country Code'], value_vars=range(2000, 2023), var_name='year', value_name='total_consumption')
renewables_percentage = renewables_percentage.melt(id_vars=['Country Name', 'Country Code'], value_vars=range(2000, 2023), var_name='year', value_name='renewables_percentage')
total_consumption.replace()

megaDF = GDP.merge(population, how='outer',on=['Country Name', 'Country Code', 'year'])
megaDF = megaDF.merge(total_consumption, how='outer',on=['Country Name', 'Country Code', 'year'])

megaDF['total_consumption_kWh'] = [i * 10**9 for i in megaDF['total_consumption']]
megaDF.drop(columns=['total_consumption'], inplace=True)

megaDF = megaDF.merge(renewables_percentage, how='outer',on=['Country Name', 'Country Code', 'year'])
megaDF.head()
```

With data on the GDP, total energy consumption, and share of total energy consumption that was renewables, we calculated GDP per capita, total energy consumption per capita, renewables consumption, and renewables consumption per capita.

```python
megaDF['GDP_per_capita'] = [megaDF['GDP'][i]/megaDF['population'][i] for i in range(len(megaDF['GDP']))]
megaDF['total_consumption_per_capita_kWh'] = [megaDF['total_consumption_kWh'][i]/megaDF['population'][i] for i in range(len(megaDF['GDP']))]
megaDF['renewables_consumption_kWh'] = [megaDF['total_consumption_kWh'][i] * megaDF['renewables_percentage'][i] for i in range(len(megaDF['GDP']))]
megaDF['renewables_consumption_per_capita_kWh'] = [megaDF['renewables_consumption_kWh'][i]/megaDF['population'][i] for i in range(len(megaDF['GDP']))]
```

Since we cannot run the models with NaNs, we started by dropping rows with nan values. Then, we checked to see which countries had data dropped, and what data was left. Oman, Saudi Arabia, Singapore, and Turkmenistan were dropped because they always had at least one column of NaNs, likely from lack of renewables data. Countries with less than ten years of data between 2000 and 2001 were dropped. This included Iceland, Iraq, and Sri Lanka.  Hong Kong, Trinidad and Tobago, and Venezuela all had at least 13 years of data left. However, the most recent year with all the data we needed from Venezuela was 2014, which we thought was not recent enough, so we dropped that as well. In the end, we had 66 countries.

```python
megaDF_model = megaDF.dropna()
print(len(megaDF['Country Name']))
print(len(megaDF_model['Country Name']))
print(len(megaDF['Country Name'])-len(megaDF_model['Country Name']))

# seeing if any countries in particular have limited data now
for i in pd.unique(megaDF['Country Name']):
    if len(megaDF_model.loc[megaDF_model['Country Name'] == i]) < 23:
        print(i + ": " +str(len(megaDF_model.loc[megaDF_model['Country Name'] == i])))

# checking to see how much recent data there is
megaDF_model.loc[megaDF_model['Country Name'] == 'Hong Kong']  # stays: recent data (2008-2022)
megaDF_model.loc[megaDF_model['Country Name'] == 'Trinidad and Tobago']  # stays: recent data (2010-2022)
megaDF_model.loc[megaDF_model['Country Name'] == 'Venezuela']  # dropped: most recent data is 2014

# dropping countries with minimal recent data (less than 10 years of data and/or last entry is more than 5 years ago)
droppedNaNs = ['Iceland', 'Iraq', 'Sri Lanka', 'Venezuela']
megaDF_model = megaDF_model[~megaDF_model['Country Name'].isin(droppedNaNs)]

megaDF_model.head()
```

# Forecasting and Prediction Modeling

In order to best answer our three prediction question, we decided to use different regression and classification models and then compare the errors of each of those (in terms of RMSE and R^2 values) to choose the best model/prediction results. The three prediction questions are as follow:
- Regression: Predict total energy consumption in the next three years of the data (2023-25)
- Regression: Predict renewable energy consumption in the next three years of the data (2023-25)
    1) Multiple Linear Regression
    2) K Nearest Neighbors
    3) Ridge Regression
    where our features are the GDP and the population, the x variable being the year and the y being the actual predicted energy consumption.
- Classification: Determine (binary) if the rate of renewable consumption will grow faster than total energy consumption in the next three years (2023-25)
    1) Bagging
    2) Boosting
    3) Random Forest

## Modeling Total Energy Consumption

MLR: We are starting by building a multiple linear regression model on the training
data and using those parameter estimates to predict the values of our target in the testing data and
training. Additionally, we will use those predictions to calculate a test RMSE and a train RMSE where the test RMSE will be used to evaluate this model's performance relative to other models we will build (Ridge, Lasso, and KNN Regression).  

```python
import warnings
warnings.filterwarnings("ignore")
megaDF_copy = megaDF_model.copy()

#errors_TEC: this dataframe will store the errors (RMSE and R^2 for training and testing) for all models, for TEC
errors_TEC = pd.DataFrame()
errors_RC = pd.DataFrame()
for country in pd.unique(megaDF_copy['Country Name']):
    new_row = {'Country Name': country}
    errors_TEC = errors_TEC.append(new_row, ignore_index=True)
    errors_RC = errors_RC.append(new_row, ignore_index=True)

#this dataframe will store the MLR, KNN, and Ridge regression for all countries 2023-25
tot_cons_predictions = pd.DataFrame()
re_cons_predictions = pd.DataFrame()
for country in pd.unique(megaDF_copy['Country Name']):
    for year in [2023, 2024, 2025]:
        new_row = {'Country Name': country, 'Year': year}
        tot_cons_predictions = tot_cons_predictions.append(new_row, ignore_index=True)
        re_cons_predictions = re_cons_predictions.append(new_row, ignore_index=True)
```

This is where we perform the MLR modeling and add the MLR predictions we get from that for both renewable and total consumption such that it will create a new column with those respective prediction results for each of the next three years for each of our countries. 

```python
mlr_pred_tot = []
test_RSME_tot = []
train_RSME_tot = []
test_R2_tot = []
train_R2_tot = []
coef_tot = []
intercept_tot = []

mlr_pred_re = []
test_RSME_re = []
train_RSME_re = []
test_R2_re = []
train_R2_re = []
coef_re = []
intercept_re = []
for country in pd.unique(megaDF_copy['Country Name']):
    X = megaDF_copy.loc[megaDF_copy['Country Name'] == country][['year', 'GDP_per_capita']]
    yTot = megaDF_copy.loc[megaDF_copy['Country Name'] == country][['total_consumption_per_capita_kWh']]
    yRE = megaDF_copy.loc[megaDF_copy['Country Name'] == country][['renewables_consumption_per_capita_kWh']]

    X_trainTot, X_testTot, y_trainTot, y_testTot = train_test_split(X, yTot, test_size=0.2, random_state=1)
    X_trainRE, X_testRE, y_trainRE, y_testRE = train_test_split(X, yRE, test_size=0.2, random_state=1)

    MLR_modelTot = LinearRegression()
    MLR_modelRE = LinearRegression()
    MLR_fit_tot = MLR_modelTot.fit(X_trainTot, y_trainTot)
    MLR_fit_re = MLR_modelRE.fit(X_trainRE, y_trainRE)

    # storing coefficient and intercept
    coef_tot.append(MLR_fit_tot.coef_)
    intercept_tot.append(MLR_fit_tot.intercept_)
    coef_re.append(MLR_fit_tot.coef_)
    intercept_re.append(MLR_fit_tot.intercept_)

    y_pred_testTot = MLR_fit_tot.predict(X_testTot)
    y_pred_trainTot = MLR_fit_tot.predict(X_trainTot)
    y_pred_testRE = MLR_fit_re.predict(X_testTot)
    y_pred_trainRE = MLR_fit_re.predict(X_trainTot)

    #error lists for curr model
    test_RSME_tot.append(mean_squared_error(y_testTot, y_pred_testTot, squared = False))
    train_RSME_tot.append(mean_squared_error(y_trainTot, y_pred_trainTot, squared = False))
    test_R2_tot.append(r2_score(y_testTot, y_pred_testTot))
    train_R2_tot.append(r2_score(y_trainTot, y_pred_trainTot))
    test_RSME_re.append(mean_squared_error(y_testRE, y_pred_testRE, squared = False))
    train_RSME_re.append(mean_squared_error(y_trainRE, y_pred_trainRE, squared = False))
    test_R2_re.append(r2_score(y_testRE, y_pred_testRE))
    train_R2_re.append(r2_score(y_trainRE, y_pred_trainRE))

    #added in the prediction for next three years 2023-25
    GDP_per_cap2022 = megaDF_copy.loc[megaDF_copy['Country Name'] == country]['GDP_per_capita'].iloc[-1]
    for year in range(2023, 2026):
        mlr_pred_tot.append(MLR_fit_tot.predict([[year, GDP_per_cap2022]])[0][0])
        mlr_pred_re.append(MLR_fit_re.predict([[year, GDP_per_cap2022]])[0][0])
        # yTot = yTot.append({'total_consumption_per_capita_kWh': MLR_fit.predict([[2023]])[0][0]}, ignore_index=True)

    # print(country, mlr_pred_tot, mlr_pred_re)  # confirming it works properly

#add in calculations and MLR preds as new columns into respective datasets
errors_TEC['MLR Test RMSE'] = test_RSME_tot
errors_TEC['MLR Train RMSE'] = train_RSME_tot
errors_TEC['MLR Test R2'] = test_R2_tot
errors_TEC['MLR Train R2'] = train_R2_tot

errors_RC['MLR Test RMSE'] = test_RSME_re
errors_RC['MLR Train RMSE'] = train_RSME_re
errors_RC['MLR Test R2'] = test_R2_re
errors_RC['MLR Train R2'] = train_R2_re

# removing 14 negative R^2 values
errors_TEC['MLR Test R2'] = errors_TEC['MLR Test R2'].apply(lambda x: np.nan if x < 0 else x)
errors_TEC['MLR Train R2'] = errors_TEC['MLR Train R2'].apply(lambda x: np.nan if x < 0 else x)

errors_TEC['MLR Coefficients'] = coef_tot
errors_TEC['MLR Intercepts'] = intercept_tot
errors_RC['MLR Coefficients'] = coef_re
errors_RC['MLR Intercepts'] = intercept_re

tot_cons_predictions['MLR Predictions'] = mlr_pred_tot
re_cons_predictions['MLR Predictions'] = mlr_pred_re
# tot_cons_predictions

# plt.scatter(XTot, yTot)
# plt.axline((0, MLR_fit.intercept_), slope=MLR_fit.coef_)

# plt.scatter(megaDF_copy['total_consumption_per_capita_kWh'], yTot)
# would be nice to make a scatter plot of all the y vals (historic) and predictions (historic and future)
# plus line y=x
```

## Classification: 

We calculated year over year changes in total energy consumption and renewable energy consumption so we could compare which was growing faster or slower per year. We then added a column that one-hot encoded if a column was country's renewable energy consumption was growing faster (1) or slower (0) than the total energy consumption. These two values were never equivalent.

```python
# adding year over year percentage changes
country = []
year = []
tot_consumption_per_cap_YOY = []
re_consumption_per_cap_YOY = []
GDP_per_cap_YOY = []
population_YOY = []
fasterSlower = []
for i in pd.unique(megaDF_model['Country Name']):
    yr = min(megaDF_model.loc[megaDF_model['Country Name'] == i]['year'])
    tot_consumption_per_cap_YOY += [np.NaN]
    re_consumption_per_cap_YOY += [np.NaN]
    GDP_per_cap_YOY += [np.NaN]
    population_YOY += [np.NaN]
    fasterSlower += [np.NaN]
    country += [i]
    year += [yr]
    while yr < max(megaDF_model.loc[megaDF_model['Country Name'] == i]['year']):
        tot_cons00 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr)]['total_consumption_per_capita_kWh'].values[0]
        tot_cons01 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr+1)]['total_consumption_per_capita_kWh'].values[0]
        
        re_cons00 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr)]['renewables_consumption_per_capita_kWh'].values[0]
        re_cons01 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr+1)]['renewables_consumption_per_capita_kWh'].values[0]
        
        GDP00 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr)]['GDP_per_capita'].values[0]
        GDP01 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr+1)]['GDP_per_capita'].values[0]

        population00 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr)]['population'].values[0]
        population01 = megaDF_model.loc[(megaDF_model['Country Name'] == i) & (megaDF_model['year'] == yr+1)]['population'].values[0]

        country += [i]
        year += [yr+1]
        tot_consumption_per_cap_YOY += [100*(tot_cons01-tot_cons00)/tot_cons00]
        re_consumption_per_cap_YOY += [100*(re_cons01-re_cons00)/re_cons00]

        # confirm this logic works even if YOY change in re or tot consumption are negative
        if re_consumption_per_cap_YOY[-1] > tot_consumption_per_cap_YOY[-1]:
            fasterSlower += [1]
        elif re_consumption_per_cap_YOY[-1] < tot_consumption_per_cap_YOY[-1]:
            fasterSlower += [0]
        # elif re_consumption_per_cap_YOY[-1] == tot_consumption_per_cap_YOY[-1]:  # this never happens
        #     fasterSlower += [2]
        GDP_per_cap_YOY += [100*(GDP01-GDP00)/GDP00]
        population_YOY += [100*(population01-population00)/population00]

        yr += 1

YOYdf = pd.DataFrame(data={'Country Name':country, 'year':year, 
'tot_consumption_per_cap_YOY':tot_consumption_per_cap_YOY, 're_consumption_per_cap_YOY':re_consumption_per_cap_YOY,
'GDP_per_cap_YOY':GDP_per_cap_YOY, 'population_YOY':population_YOY, 'fasterSlower':fasterSlower})

megaDF_model = megaDF_model.merge(YOYdf, how='outer',on=['Country Name', 'year'])
megaDF_model = megaDF_model.dropna()

print(sum(megaDF_model['fasterSlower']))
print(sum(megaDF_model['fasterSlower'])/len(megaDF_model['fasterSlower']))

megaDF_model.head()
```
